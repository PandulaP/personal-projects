# Medical Lasers Anomaly Detection

## Introduction


For quality assurance purposes, medical laser manufacturers want to implement systems which are capable of recognizing defective products among the rest. Ideally, properly working medical laser products are supposed to produce a **constant light output with a frequency as constant as possible**, thus only certain *(small)* fluctuations are accepted. Therefore, any implemented system should be able to identify any laser product whose **power fluctuates to an intolerable extent**.

Below figure provides examples of measurement series of correctly working lasers (left) and faulty lasers (right). The respective measurements thus describe which type of fluctuations are or are not tolerable.  

![title](laser_img.JPG)

---
In this project, I plan on developing predictive model(s) based on the given set of laser intensity measurement data which can correctly identify the functioning of a laser based on respective laser intensity measurements.


## Background on the problem

Considering the nature of this problem, I identify this to fall under broad domains of **time series classification** and **pattern recognition** fields. 

* Even more so, this can be treated as an overlap of the above mentioned fields due to tha fact that, the measurement data/attributes at our displosal are *time-series* and the whole process would rely on being able to develop a machanism which is capable of identifying abnormal patterns/behaviors over these sequential data-points.
* In other words, our end goal would be to use these sequential data-points as features and develop a model that can decide whether the source of each considered measurement series, i.e., the corresponding medical laser product is suitable for sale or not, so that we can classify each medical laser product as **good** or **bad**.

## Making an initial hypothesis:

* After going through the problem description and and by simply observing the demostrated plots, we can realize that, whatever the model that we develop to identify good vs. bad laser products should put its sole focus on **being able to differentiate the varying power fluctuations** of a bad laser product from a good product.

* Thus, prior moving to solve the problem at hand, I first make below hypothesis about the problem using all available information and try to validate these hypothesis during the initial stages of the project, mainly by performing EDA on the available data.

My hypotheses are;
* bad laser prouduct (not suitable for sale) : showcase **high power fluctuations** *(i.e., highly volatile power intensity readings over monitored timeframe)*
* good laser prouduct (suitable for sale) : showcase **low power fluctuations** *(i.e., almost constant power intensity readings over monitored timeframe)*

### Creating new features to highlight distinct characteristics:

As discussed above, the aim of feature-engineering would be to create new features that would allow any pedictive model to easily identify a sellable product from a non-sellable product simply by the characteristics of respective measurement data.

Thus these **new features** should focus on **highlighting the facts which are unique** to either of the two classes.
<br><br>
* Considering the above idealogies and observations, we can perform below transformations on the measurement data and do variable creations hoping to yeild better results at the modeling stage:
<br><br>
    * `sd_m` : **standard deviation measurement data** — we simply take the **standard deviation** of the intensity measurement data for each product. 
<br><br>
    * `m_diff` : **generate the stationary measurement time-series for product** — here we create a the new stationary time series for each product using the **first order-differencing** method.
        * `m_diff` at (t) = measurement at (t) - measurement at (t-1)
<br><br>
    * `sd_m_diff` : **standard deviation of `m_diff`** — this is the **standard deviation** of the newly created stationary time series (`m_diff`) for each product. 
<br><br>
    * `max_m_diff` : **max of `m_diff`** — the **maximum** value of the newly created stationary time series for each product.
<br><br>
    * `sd_m_5sec_roll_max` : **maximum of the 5 second rolling standard deviations of measurement data** — this statistic is generated by computing **standard deviation** for each **4-seconds** time period for the measurement data and then taking the **maximum** of all those standard-deviation values.
<br><br>
    * `moving_avg_m_diff_2sec_max` : **overall maximum of the 2-sec moving average values generated on the newly created stationary time series** — here a 2-second moving average is calculated on `m_diff` series and then the maximum of all moving average values are taken for each product.
<br><br>
    * `moving_avg_m_diff_2sec_min` : **overall minimum of the 2-sec moving average values generated on the newly created stationary time series** — similar to above, the minimum of the 2-second moving average values calculated on `m_diff` data is taken for each product.
<br><br>
    * `moving_avg_m_diff_2sec_diff` : **absolute difference between `moving_avg_m_diff_2sec_min`and `moving_avg_m_diff_2sec_max` features**
<br><br>

Below noted are few additional features which were **created, <u>but dropped</u>** due to not showing any meaningful separation between two product groups. 

   * `median_m_diff` : **meadian** of the newly created stationary time-series `m_diff`
   * `moving_avg_m_3sec_max` : **overall maximum** of the 3-sec moving average values generated on the original measurement data for each product.

Now having above features at hand, which were created with the **sole purpose of introducing separability to the two product groups within the feature-space**, I believe it would be fine to **discard the original measurement data** and **only move forward with these new set of features** to modeling stage.

And knowing that we will have to perform the same set of feature-engineering steps on the **test-set** when measuring model performance, I built a function that contains all above feature-engineering steps as it will be convinient for us to re-use.

```python
def create_new_features(dataset):

    # generate the stationary time series `tmp_df`
    tmp_df = dataset.diff(periods=1,axis=1).abs().copy()
    nw_cl_names = ['m_diff_'+ str(col) for col in dataset.columns]
    tmp_df.columns = nw_cl_names
    tmp_df.drop('m_diff_1 sec',axis = 1, inplace=True) # since first column will always be `NaN`
    
    tmp_df['sd_m'] = dataset.std(axis=1)    
    
    tmp_df['sd_m_diff'] = tmp_df.std(axis=1)
    
    tmp_df['max_m_diff'] = tmp_df.max(axis=1)
    
    tmp_df['sd_m_5sec_roll_max'] = dataset.rolling(5,axis=1).std().max(axis=1)
    
    tmp_df['moving_avg_m_diff_2sec_max'] = tmp_df.rolling(2,axis=1).mean().max(axis=1)
    
    tmp_df['moving_avg_m_diff_2sec_min'] = tmp_df.rolling(2,axis=1).mean().min(axis=1)
    
    tmp_df['moving_avg_m_diff_2sec_diff'] = np.abs(tmp_df['moving_avg_m_diff_2sec_min'] - tmp_df['moving_avg_m_diff_2sec_max'])    
        
    tmp_df_reduced = tmp_df.iloc[:,-7:] # dataset only with `std`, `max` and `median` statistics
    
    return tmp_df_reduced, tmp_df
```

As you can see, here the fucntion generates **two datasets**, one which has all the newly created variables and the other excluding the stationary time-series data-points (thus only contains the *summary statistic values* calculated based on the measurement data of each product)

```python
# generate feature-engineered training datasets
X_train_new_red, X_train_new = create_new_features(X_train)
```

## Modeling

During the modelling stage, I will be considering all or different combinations of the above engineered features to identify a hyperplane that can successfully separate the two product groups.

* Before moving to create any advanced models, I believe it is best that we create a simple **baseline** model and start with a simple hypothesis space first.

But knowing that **logistic regression** models has always been treated as the "go-to" method for most binary-classification problems, I plan on first implementing a simpler LR model on the dataset and use it has the base-line model for any more sophisticated models that I would build later on.

Now recalling what we observed during the EDA process, *i.e., not seeing a clear seperation of **sellable** and **non-sellable** products on the feature space*, I believe the problem at hand would **mostly likely have a non-linear decision boundry**

Therefore, having the above results and considering it as a base-line, as the next step I train a `SVM` with a `Gaussian` kernel hoping that it would yeild to better performance.

### 2. Fitting a `SVM` classifier with `Gaussian` Kernel

* Since now that we have a base-line model to compare against with, our aim here would be to develop a model which has a higher predictive power than the baseline.
* As we know, `SVM`s become extremely powerful when they are combined with **higher dimensional kernels**.
* These kernels project our data into higher dimensional spaces so that we are able to fit non-linear relationships with a linear classifier on those higher dimensional spaces.

It's possible for us to fit a `Gaussian` kernel by assigning `RBF` *(radial basis function)* for our `kernel` hypyerparameter *(note that Gaussian kernel is a special case of RBF)*

* As we saw above, there appears to be some amount of over-lap between the data points coming from each class and to handle this, we can tune the value of hyperparameter `C` and allow some data points to fall into the margin if that allows a better fit.

* The hyperparameter `C` corresponds to the **hardness** of the margin, thus altering this value would allow the algorithm to be less strict on setting up the margin and allow leakages and let cross overs happen by setting up smaller-margin hyperplanes with large enough `C` values.

Said above, the optimal `C` value depends on the dataset, thus I will try out different values and decide the best for our use-case by employing the **grid search cross-validation** which can be applied through `GridSearchCV` method in `sklearn`. And In doing so, I will adjust `C` *(margin hardness)* and `gamma` values *(controls the size of the rbf kernel)* to determine the best model.

#### 2.1 Does our data align with SVM model assumptions?

Prior we move on to fit SVM classifier on our feature-data, it's important to understand its requirement of expecting feature data in a standard range.

* SVMs assume that the feature data we use are in a standard range and therefore, normalization of the feature-data prior feeding them to the SVM model is always recommended.

* Mentioned that, I believe if we normalize the engineered features, it will result with a loss of information due to the fact that;
    * first we are dealing with **sequential data** and,
    * secondly, the whole idea of feature engineering in this particular problem was to **highlight abnormal patterns that exists** within the measurement data.


Therefore, I divide this section into two sub-sections where I try fitting SVMs with and without feature-scaling and access the classification power of each approach.

* First let's have a look at current distributions of our feature variables.

```python
X_train_new_red.hist(figsize=(15,10))
plt.tight_layout()
```

It's clear that the features are not having a standard scale. Now let's **standerdize** all these them and **grid search** for the best model parameters as described before:
*(Note that I decide to **Standerdize instead of Normalize** because I would like to preserve any outlier behaviors that exist in features - I believe the normalization process will reduce the impact coming from extreme values)*

```python
# Standerdize the feature-space
standard_scalr = StandardScaler()
X_train_new_red_scaled = standard_scalr.fit_transform(X_train_new_red)
```

```python
pd.DataFrame(X_train_new_red_scaled,columns = X_train_new_red.columns).hist(figsize=(15,10))
plt.tight_layout()
```

#### 2.2 SVM on scaled data :

* Now having the above scaled features in hand, let's perform **grid search** and assess the resuting model:

```python
# parameters to grid search
parameter_grid = {'C': np.arange(0.01,20,0.05),
                  'gamma': np.arange(0.01,10,0.1)}
```

```python
#create GridSearch object with SVM classifier
svm_grid = GridSearchCV(SVC(kernel='rbf'), parameter_grid ,cv=3, n_jobs= -1)
```

```python
# fit the grid search model on scaled data
%time svm_grid.fit(X_train_new_red_scaled, y_train.values.ravel())
```

* After the grid search completes, we can find the **optimal** hypterparameter values for `gamma` and `C` as below:

```python
C_opt = svm_grid.best_params_['C']
gamma_opt = svm_grid.best_params_['gamma']

print(f'Optimal hyperparameter values for SVM model;\n - C : {C_opt}\n - gamma : {gamma_opt}')
```

Now in order to measure the predictive power of the above trained model, let's apply the same transformation we did on our training feature set on the test set and generate the predictions

```python
y_pred_svm_opt_scaled = svm_grid.predict(standard_scalr.transform(create_new_features(X_test)[0]))
```

```python
print(classification_report(y_test, y_pred_svm_opt_scaled))
```

```python
cf_mat_2 = confusion_matrix(y_test,y_pred_svm_opt_scaled)
sns.heatmap(cf_mat_2.T, square=True, annot=True, fmt='d', cbar=False,
            xticklabels=['non-sellable','sellable'],
            yticklabels=['non-sellable','sellable'])
plt.xlabel('true label')
plt.ylabel('predicted label')
plt.title('Confusion Matrix for SVM with Gaussian Kernel - Scaled Feature Data\n')
plt.show()
```

**When comparing above results against the base-line:**

   * **Accuracy** has improved by **3%** *(86% of all sellable and non-sellable products are correctly predicted)*
   
   * For **non-sellable** products:
       * **Precision** has improved by **5%** *(83% of all predicted non-sellable products are truely non-sellable)*

### 3. Fitting a `Random Forest` classifier

* Since Random Forest classifier tend to perform really well with the default hyperparameter values, before moving in to tuning each individual hyperparameter, here I first fit a base-line classifier and measure it's predictive capability.

* Afterwards, in order to derive the best possible model for our problem, I am employing a mix of **validation curves** and the `GridSearchCV` method in a similar way to what was done for the `Gaussian SVM`.

##### Why combine **validation curves** with `GridSearchCV`?

* The reason for not blindly inserting set of parameter values for the `GridSearchCV` method for random forest is that, as it is seen below, even though there were only **2** hyperparameters to tune in `SVM` classifier, `Random Forest` classifier has **much more**, thus if we employ the grid search method with a random and a broad set of candidate hyperparameter values to test for optimality, it will become computationally extremely heavy and take forever to find the best model.

* Instead, here we will first try to optize each hyperparameter separately (i.e., optimize one hyperparameter value while keeping the remaining at their defaults) and then identify a small set of potentially optimum hyperparameter values to test on `GridSearchCV` method.

#### 3.1. Base-line `Random Forest` classifier

```python
rf_bl_mod = RandomForestClassifier(random_state=16)

rf_bl_mod.fit(X_train_new_red, y_train.values.ravel())
```

* Similarly, other performance metrics *(precision, recall and f1-score)*  record better values in general than of the base-line model *(overall in predicting both classes)*.

#### 3.2. Optimize hyperparameters with **Validation Curves** *(for Accuracy)*

* In this approach, I will visually check for potential optimum values for each of the `RandomForestClassifier`'s hyperparameters separately.


* **Validation curve** is a plot which displays how the model **accuracy** performs at different values of a single hyperparameter.


* Through-out the rest of this part, we will be trying to optimize the below set of hyperparameters of the `RandomForestClassifier`:

    * `n_estimators` = number of trees in the foreset
    * `max_features` = max number of features considered for splitting a node
    * `max_depth` = max number of levels in each decision tree
    * `min_samples_split` = min number of data points placed in a node before the node is split
    * `min_samples_leaf` = min number of data points allowed in a leaf node


* As it's seen, even though there are few other hyperparameters which we can optimize the classifier upon, I am not trying to tune all of them since I believe the above to have the most affect on model's predictive performance.

```python
def generate_validation_curve(classifier,p_name,p_range,x_tr,y_tr,r_state):

    train_scores, test_scores = validation_curve(classifier(random_state =r_state),
                                                     X = x_tr, y = y_tr,
                                                     param_name = p_name,
                                                     param_range = p_range,
                                                     cv = 3)

    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)

    plt.figure(figsize=(8,5))
    plt.title(f"Validation Curve with '{p_name}' hyperparameter\n")
    plt.xlabel(f"{p_name}")
    plt.ylabel("Score")
    plt.ylim(0.0, 1.1)
    lw = 1
    plt.plot(p_range, train_scores_mean, label="Training score", color="red", lw=lw)
    plt.fill_between(p_range, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.2,
                     color="red", lw=lw)
    plt.plot(p_range, test_scores_mean, label="Cross-validation score",
                 color="blue", lw=lw)
    plt.fill_between(p_range, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.2,
                     color="blue", lw=lw)
    plt.legend(loc="best")
    plt.show()
```

```python
x_tr, y_tr  = X_train_new_red, y_train.values.ravel()
```

##### a) `n_estimators`

```python
generate_validation_curve(RandomForestClassifier, 'n_estimators', np.arange(1,15,1), x_tr,y_tr,16)
```

##### What does the curve say?

* If both train and validation scores are **low**, model is **underfitting**.
* If **train** score is **high** and **validation score** is **low**, model is **overfitting**.
* Values where scores are **high** for both **train** and **validation** can be considered as good candidate values. 

**What value range to use for `n_estimators`?**
  * As we can see, after `n_estimators` exceed **5**, the model seems to **overfit**.
  * Therefore, lets consider **<u>1 to 10</u>** range as a candidate range for the `n_estimator`.
  
Similar to above, I will decide candidate value ranges for the rest of the hyperparameters as below:

##### b) `max_features`

```python
generate_validation_curve(RandomForestClassifier, 'max_features', np.arange(1,8,1), x_tr,y_tr,16)
```

* Planning to consider **<u>2 to 6</u>** range for the `max_features`.

##### c) `max_depth`

```python
generate_validation_curve(RandomForestClassifier, 'max_depth', np.arange(1,8,1), x_tr,y_tr,16)
```

* Planning to consider **<u>2 to 5</u>** range for the `max_depth`.

##### d) `min_samples_split`

```python
generate_validation_curve(RandomForestClassifier, 'min_samples_split', np.arange(2,15,1), x_tr,y_tr,16)
```

* Planning to consider **<u>4 to 12</u>** range for the `min_samples_split`.

##### e) `min_samples_leaf`

```python
generate_validation_curve(RandomForestClassifier, 'min_samples_leaf', np.arange(1,10,1), x_tr,y_tr,16)
```

* Planning to consider **<u>3 to 9</u>** range for the `min_samples_leaf`.

#### 3.3. Derive the optimal `Random Forest` classifier with `GridSearchCV ` method:

## DISCUSSION :  `SVM` vs. `Random Forest` : is one better than the other?

* As of now, we have **2 models** in hand with the best predictive power. Let's analyze their performance metrics below:

```python
def generate_summary():

    fig, ax = plt.subplots(ncols=2,nrows=1,figsize=(12,8))

    sns.heatmap(cf_mat_3.T, square=True, annot=True, fmt='d', cbar=False,
                xticklabels=['non-sellable','sellable'],
                yticklabels=['non-sellable','sellable'], ax = ax[0])
    ax[0].set_xlabel('true label')
    ax[0].set_ylabel('predicted label')
    ax[0].set_title('SVM classifier on Unscaled Data\n')

    sns.heatmap(cf_mat_4.T, square=True, annot=True, fmt='d', cbar=False,
                xticklabels=['non-sellable','sellable'],
                yticklabels=['non-sellable','sellable'], ax = ax[1])
    ax[1].set_xlabel('true label')
    ax[1].set_ylabel('predicted label')
    ax[1].set_title('Random Forest Classifier\n')

    fig.show()
    print(f'classification report for SVM on unscaled data:\n\n{classification_report(y_test, y_pred_svm_opt)}\n\nclassification report for Random Forest:\n\n{classification_report(y_test, y_pred_rf_opt_mod)}')
```

##### When we compre the predictive power of both models over the **Test** set:

* **Overall accuracy** is **slightly higher** *(~1%)* in **SVM**.
* `f1-score` is also slightly higher for both classes *(sellable & non-sellable product)* in **SVM**.
    * Therefore, I believe it's safe to arrive at the conclusion that, overall, the **SVM** trained on **un-scaled feature data** tend to perform best for the considered classification task.
    
    
###### If correctly predicting non-sellable products is more important:
* `SVM` tend to suite better as it has a higher **recall**
* But at the same time, if wrongly classifying a sellable product as non-sellable is costly, `Random Forest` would be better since it has **100% precision** in classifying non-sellable products.


###### If correctly predicting sellable products is more important:
* `Random Forest` tend to suite better as it has a perfect **recall** of 100%.
* Similar to above, if classifying a non-sellable product as sellable is costly, the `SVM` would be better due to the **slightly high precision** *(~90%)* it has in classifying sellable products.


Therefore, it is evident that the **selection of model should depend on the use-case** but not solely on the predictive performance of the model.

If no preference is made on requiring to correctly predict one class over the other, **SVM with Gaussian kernel** *(trained on the un-scaled feature data)* can be recommended over the random forest classifier.
